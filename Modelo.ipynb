```python

import pandas as pd
import numpy as np

from xgboost import XGBRegressor
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier
from xgboost.sklearn import XGBClassifier
from sklearn.model_selection import train_test_split
from lightgbm import LGBMClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.gaussian_process.kernels import RBF
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
```

```python
url = "dataset.csv"
x = pd.read_csv(url)
x.info()
```

```python
x.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
x.head()
```

```python
y = x['EVOLUCAO']
x.drop(['EVOLUCAO'], axis=1, inplace=True)

x_train, x_test, y_train, y_test = train_test_split(x, y)
```

```python
classificadores = [ 
    RandomForestClassifier(),
    XGBClassifier(),
    LGBMClassifier(),
    SVC(),
    DecisionTreeClassifier(),
    AdaBoostClassifier(),

]
nomes = [
    "Random Forest",
    "XGBRegressor",
    "Light GBM",
    "RBF SVM",
    "Decision Tree",
    "AdaBoost"
]
```

```python
for classificador, nome in zip(classificadores, nomes):
    score = cross_val_score(classificador, x, y, cv =10)
    print(f"Accuracy of {nome} Classifier is:{score}")
    print('MÃ©dia: {:} | Desvio: {:.2}'.format(np.mean(score), np.std(score)))
```
